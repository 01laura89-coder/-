Тапсырма 1: CSV Файлын Оқу және Тізімге Сақтау

```python
with open("/content/Salaries.csv", "r", encoding="utf-8") as file:
  lines = file.readlines()
data_rows = []
for line in lines[1:11]:
  data_rows.append(line.strip().split(","))
print("Жол саны:", len(data_rows))
print ("Бірінші қызметкердің дерегі:", data_rows[0])
```

Тапсырма 2: Кортеж және Сөздік – Алғашқы 3 Адамның Атын Сақтау

```python
first_three_names = (data_rows[0][1], data_rows[1][1], data_rows[2][1])
print(first_three_names)
```

Тапсырма 3: Тізімдік Өрнектер (List Comprehension)

```python
base_pays_str = [row[3] for row in data_rows]
print(base_pays_str)
```

Тапсырма 4: Жиын (Set) – Бірегей Атақтарды Табу

```python
JobTitle = []
for row in data_rows:
  JobTitle.append(row[2])
unique_titles = set(JobTitle)
print(JobTitle)
```
